```python
#### IMPORT STATEMENTS ####

# None

#### FUNCTION DEFINITIONS ####

#1. Define what car_classification they used and how much it was per day (B, S, and L)
car_packages = {
    "B": {"classification": "Budget", "daily_cost": 40.00},
    "S": {"classification": "Standard", "daily_cost": 60.00},
    "L": {"classification": "Luxury", "daily_cost": 80.00}
    }
#2. Have the function display the full classification, not just a letter.
def get_car_classification(package_letter):
    """
    Returns the full classification name for a given car package letter.
    """
    if package_letter in car_packages:
        return car_packages[package_letter]['classification']
    else:
        return "Invalid package letter"

#3. How many days was the car rented for
def get_rental_days():
    """
    Prompts the user for the number of rental days and validates the input.
    Returns a positive integer for the number of rental days.
    """
    while True:
        try:
            rental_days = int(input("Enter the number of rental days: "))
            if rental_days > 0:
                return rental_days
            else:
                print("Number of days must be a positive number.")
        except ValueError:
            print("Invalid input. Please enter a whole number for days.")

#4. What was the starting miles, and final miles
def get_miles():
  starting_miles = float(input("Enter the starting miles: "))
  final_miles = float(input("Enter the final miles: "))
  return final_miles - starting_miles
#5. The first 100 miles are free, every mile afterwards is $0.35 per mile
def calculate_cost(miles):
  if miles <= 100:
    return 0
  else:
    # Calculate cost for miles over 100
    return (miles - 100) * 0.35
#6. What is the total cost?
def calculate_total_cost(car_package_choice, rental_days, miles):
    daily_cost = car_packages[car_package_choice]['daily_cost']
    rental_cost = daily_cost * rental_days
    mileage_cost = calculate_cost(miles)
    return rental_cost + mileage_cost

#### PROGRAM LOGIC ####

#1. Ask the customer for what car package (B, S, or L)
while True:
    car_package_choice = input("Enter car package (B for Budget, S for Standard, L for Luxury): ").upper()
    if car_package_choice in car_packages:
        break
    else:
        print("Invalid car package. Please enter B, S, or L.")

#2. Ask the customer for how many days
rental_days = get_rental_days()

#3. Input how many miles was the final miles
miles = get_miles()

# Calculate and print total cost
total_cost = calculate_total_cost(car_package_choice, rental_days, miles)
print(f"\nCar Classification: {get_car_classification(car_package_choice)}")
print(f"Rental Days: {rental_days}")
print(f"Miles Driven: {miles}")
print(f"Total Cost: ${total_cost:.2f}")
```
