```python
#### IMPORT STATEMENTS ####
        # None
#### CLASS DEFINITIONS ####
class Book:
    def __init__(self, title, author, is_available=True):
  # Name of Book
      self.title = title
  # Name of the Author
      self.author = author
  # Variable to know if it can be checked out or not
      self.is_available = is_available

    def __str__(self):
  # Return the information all together in a string
        status = "Available" if self.is_available else "Checked Out"
        return f"'{self.title}' by {self.author} ({status})"

    def checkout(self):
# Checkout Method, to determine if its checked out or available.
      if self.is_available:
          self.is_available = False
          current_status = "Checked Out"
          print(f"'{self.title}' has been checked out.")
      #e Else, You can check it out
      else:
          self.is_available = True
          current_status = "Available"
          print(f"'{self.title}' is already checked out.")

    def return_book(self):
      if not self.is_available:
          self.is_available = True
          print(f"'{self.title}' has been returned.")
          return True
      else:
          print(f"'{self.title}' is already available.")
          return False

#### PROGRAM LOGIC ####


first_book = Book("The Hobbit", "J.R.R Tolkien", True)
print(first_book)

second_book = Book("Think Python", "Allen Downey", True)
print(second_book)

first_book.checkout()
```
