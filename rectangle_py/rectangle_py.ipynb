```python
#### IMPORT STATEMENTS ####
import math
#### FUNCTION DEFINITIONS ####
# Define a function that takes length and width
# Define calculate_rectangle - Area (length * width); Perimeter (2 * (length + width))
def calculate_rectangle(length, width):
    # Calculate Area
    Area = length * width
    # Calculate Perimeter
    Perimeter = 2 * (length + width)
    # Returns the Area and Perimeter
    return Area, Perimeter
#### PROGRAM LOGIC ####

# Ask the user to input the length and the width
length_input = float(input("Enter the length: "))
width_input = float(input("Enter the width: "))
# Call the calculate_rectangle function
calculated_area, calculated_perimeter = calculate_rectangle(length_input, width_input)
# Print the results from the function called
print(f"The area of the rectangle is: {calculated_area}")
print(f"The perimeter of the rectangle is: {calculated_perimeter}")
```
