```python
### IMPORT STATEMENTS ###
import pandas as pd
from datetime import datetime

### FUNCTION DEFINITIONS ###
# Make the empty list for the transactions.
transaction = []

# Add a transaction entry in (The record for the Date, Amount, Category, Vendor, and Description)
def add_transaction(date, amount, category, vendor, description):
    # Create a dictionary for the new transaction
    new_transaction = {
        "Date": date,
        "Amount": amount,
        "Category": category,
        "Vendor": vendor,
        "Description": description
    }
    # Add the new transaction to the global list of transactions
    transaction.append(new_transaction)
    print(f"Transaction added: {date} - ${amount:.2f} - {category} - {vendor} - {description}")

def get_total_expenses():
    total_expenses = sum(t['Amount'] for t in transaction)
    return total_expenses

def get_total_by_category(category_name):
    category_total = sum(t['Amount'] for t in transaction if t['Category'].lower() == category_name.lower())
    return category_total

def get_total_by_vendor(vendor_name):
    vendor_total = sum(t['Amount'] for t in transaction if t['Vendor'].lower() == vendor_name.lower())
    return vendor_total

### PROGRAM LOGIC ###
# Ask the user what they would like to pick from on the list (1. Add New Expense, 2. Get Totals, 3. Exit)
while True:
  print("What would you like to do?")
  print("1. Add New Expense")
  print("2. Get Totals")
  print("3. Exit")
  user_input = input("Your Choice: ").upper()

  if user_input == "3":
    print("Have a great day! Enjoy your Expenses")
    break
  # If the User selects "Add New Expense" call the "add_transaction" Function
  elif user_input == "1":
    date_str = input("Enter the date (YYYY-MM-DD): ")
    while True:
      amount_str = input("Enter the amount: ").replace('$', '')
      try:
        amount = float(amount_str)
        break
      except ValueError:
        print("Invalid amount. Please enter a valid number.")
    category = input("Enter the category: ")
    vendor = input("Enter the vendor: ")
    description = input("Enter a description: ")
    add_transaction(date_str, amount, category, vendor, description)
    print("Transaction added successfully!")
  # If the User puts in the "Get Totals" Option, Put them into a sub-menu (A: All Expenses, C:Filter by Category, V: Filter by Vendor)
  elif user_input == "2":
    print("What would you like to filter by?")
    print("A. All Expenses")
    print("C. Filter by Category")
    print("V. Filter by Vendor")
    sub_user_input = input("Your Choice: ").upper()

    # If the user inputs "A" into the sub-menu, Show all Transactions and Totals
    if sub_user_input == "A":
      if not transaction:
        print("No transactions recorded yet.")
      else:
        print("\n--- All Expenses ---")
        for i, t in enumerate(transaction):
          print(f"{i+1}. Date: {t['Date']}, Amount: ${t['Amount']:.2f}, Category: {t['Category']}, Vendor: {t['Vendor']}, Description: {t['Description']}")
        total_expenses = get_total_expenses()
        print(f"\nTotal Expenses: ${total_expenses:.2f}")
        print("\n--- Raw Transaction List (for testing) ---")
        print(transaction)
    # If the user inputs "C" into the sub-menu, to filter by "Category"
    elif sub_user_input == "C":
        filter_category = input("Enter category to filter by: ")
        filtered_transactions = [t for t in transaction if t['Category'].lower() == filter_category.lower()]
        if not filtered_transactions:
            print(f"No transactions found for category: {filter_category}")
        else:
            print(f"\n--- Expenses for Category: {filter_category} ---")
            for i, t in enumerate(filtered_transactions):
                print(f"{i+1}. Date: {t['Date']}, Amount: ${t['Amount']:.2f}, Vendor: {t['Vendor']}, Description: {t['Description']}")
            category_total = get_total_by_category(filter_category)
            print(f"\nTotal for {filter_category}: ${category_total:.2f}")
            print("\n--- Raw Transaction List for Category (for testing) ---")
            print(filtered_transactions)

    # If the user inputs "V" to filter by "Vendor"
    elif sub_user_input == "V":
        filter_vendor = input("Enter vendor to filter by: ")
        filtered_transactions = [t for t in transaction if t['Vendor'].lower() == filter_vendor.lower()]
        if not filtered_transactions:
            print(f"No transactions found for vendor: {filter_vendor}")
        else:
            print(f"\n--- Expenses for Vendor: {filter_vendor} ---")
            for i, t in enumerate(filtered_transactions):
                print(f"{i+1}. Date: {t['Date']}, Amount: ${t['Amount']:.2f}, Category: {t['Category']}, Description: {t['Description']}")
            vendor_total = get_total_by_vendor(filter_vendor)
            print(f"\nTotal for {filter_vendor}: ${vendor_total:.2f}")
            print("\n--- Raw Transaction List for Vendor (for testing) ---")
            print(filtered_transactions)
```
