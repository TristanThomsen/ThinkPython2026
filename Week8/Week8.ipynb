```python
#### Import Statements ####

# None

#### Function Definitions ####

#1. Define Palindrome
def is_palindrome(word):
    word_lower = word.lower()
    if word_lower == word_lower[::-1]:
        return True
    else:
        return False
#2. Return Value
def main():
    word = input("Enter a word: ")
#3. Case Insensitivity
    word = word.lower()
#4. Reversed Logic
    reversed_word = word[::-1]
#5. Program Logic
    if word == reversed_word:
        print(f'"{word}" is a palindrome.')
    else:
        print(f'"{word}" is not a palindrome.')


#### Program Logic ####

#1. Ask the user for a word
user_word = input("Enter a word to check if it's a palindrome: ")
#2. Check if the word they enter, is a palindrome
is_user_word_palindrome = is_palindrome(user_word)
#3. Return the Result through Print
print(f"The word '{user_word}' is palindrome: {is_user_word_palindrome}")
```
