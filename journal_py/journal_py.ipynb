```python
# Create a Program that records journal entries and saves them to a .txt file!

#### IMPORT STATEMENTS ####
import json
journal_file = "journal.txt"
#### FUNCTION DEFINITIONS ####

# Function 1: (write_entry): * This function should accept a filename and a string (the entry) as parameters.
def write_entry(filename, entry):
  # It must open the file in append mode ('a") for writing
  with open(filename, 'a') as file:
    file.write(entry + '\n')
  # The entry should be written into the file, each entry starting on a new line.

# Function 2: (read_journal):
def read_journal(filename):
  # Should be open in ('r') read mode
  try:
    with open(filename, 'r') as file:
      content = file.read()
      if content:
        print("\n--- Your Journal Entries ---")
        print(content.strip())
        print("----------------------------\n")
      else:
        print("\nJournal is empty.\n")
  except FileNotFoundError:
    print("\nJournal file not found. Start by adding an entry.\n")
  # Should read all the contents entered into the jounral.

# To close to the program out, enter 'Quit'



#### PROGRAM LOGIC ####
# Ask the User what do they wanna do? ("A"dd, "R"ead or "Q"uit)
while True:
  print("What would you like to do?")
  print("A - Add a new entry")
  print("R - Read all entries")
  print("Q - Quit")
  user_input = input("Your choice: ").upper()
  if user_input == "Q":
    print("Have a great day! Bye!")
    break
# If they choose 'A', ask for their thoughts and call write_entry.
  elif user_input == "A":
    entry = input("Enter your Journal Entry: ")
    write_entry(journal_file, entry)
    print("Entry added successfully!")
# If they choose 'R', call read_journal.
  elif user_input == "R":
    read_journal(journal_file)
  else:
    print("Invalid choice. Please try again.")
# Wrap this in a while loop so the user can keep using the app until they type "Quit".
```
