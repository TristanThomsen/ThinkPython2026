```python
#### IMPORT STATEMENTS ####
  # None

#### FUNCTION DEFINITIONS ####
# Create a function that will add students to the dictionary (add_students)
def add_students(gradebook_dict, name, grade):
    gradebook_dict[name] = grade

# Create a fruitful function that calculates the average (calculate_average) that shows the average of all the students grades.
def calculate_average(gradebook_dict):
    if not gradebook_dict: # Handle case of empty gradebook to avoid ZeroDivisionError
        return 0
    total_grade = sum(gradebook_dict.values())
    average_grade = total_grade / len(gradebook_dict)
    return average_grade

#### PROGRAM LOGIC ####
# Make an empty gradebook
gradebook = {}

# Use a while loop, to allow you to enter their name and grade into the gradebook.
while True:
  name = input("Enter student's name (or 'Done' to finish): ")
  if name.lower() == 'done':
    break
  try:
    grade = float(input(f"Enter {name}'s grade: "))
    add_students(gradebook, name, grade)
  except ValueError:
    print("Invalid grade. Please enter a number.")

# Enter "Done" when finished
# Call add_students to populate the dictionary

# Calculate the average grade
average = calculate_average(gradebook)
print(f"Class average: {average:.2f}")

# Print a final report showing each student's name and their grade.
print("\nFinal Report:")
for name, grade in gradebook.items():
    print(f"{name}: {grade}")
```
