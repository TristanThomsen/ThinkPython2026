```python
# IMPORT STATEMENTS
# (None)

# FUNCTION DEFINITIONS
# 1. Define verify_pin():
def verify_pin():
  #   - Create a loop that asks for a pin until it equals "1234"
  while True:
    pin = input("Enter your pin: ")
    if pin == "1234":
      return True
    else:
      print("Incorrect PIN. Please try again.")
# - Return True (This is handled by the return True inside the loop) -
# Otherwise, return False (This function is designed to loop indefinitely until
# the correct PIN is entered, thus always returning True upon successful exit of
# the loop.)

# 2. Define withdraw(balance, amount):
def withdraw(balance, amount):
  #   - Check if enough money exists
  if amount > balance:
    print("Insufficient funds.")
    return balance # Return original balance if insufficient funds
  else:
    #   - Return the new (or old) balance
    return balance - amount

# PROGRAM LOGIC
# 1. Set current_balance to 500
current_balance = 500
# 2. Call verify_pin()
if verify_pin():
  print("Access granted.")
else:
  print("Access denied.")
# 3. Get withdrawal amount from user input
amount = float(input("Enter withdrawal amount: "))
# 4. Update current_balance by calling withdraw()
current_balance = withdraw(current_balance, amount)
# 5. Print the final balance
print("Final balance:", current_balance)
```
