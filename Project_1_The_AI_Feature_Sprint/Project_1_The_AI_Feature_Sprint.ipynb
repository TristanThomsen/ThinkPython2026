```python
COMPLETED_TASKS_FILE = 'completed_tasks.txt'
tasks = []

def add_task():
    task_name = input("Enter task name: ").strip()
    if not task_name:
        print("Task name cannot be empty. Please try again.")
        return

    while True:
        try:
            priority = int(input("Enter priority (1: High, 2: Medium, 3: Low): "))
            if 1 <= priority <= 3:
                break
            else:
                print("Priority must be 1, 2, or 3. Please try again.")
        except ValueError:
            print("Invalid input. Please enter a number for priority.")

    priority_map = {1: 'High', 2: 'Medium', 3: 'Low'}
    tasks.append({'name': task_name, 'priority': priority, 'priority_text': priority_map[priority]})
    print(f"Task '{task_name}' with priority '{priority_map[priority]}' added.\n")

def list_tasks():
    if not tasks:
        print("No tasks to display.\n")
        return

    # Sort tasks by priority (lower number = higher priority)
    sorted_tasks = sorted(tasks, key=lambda x: x['priority'])

    print("--- Your Tasks ---")
    for i, task in enumerate(sorted_tasks):
        print(f"{i+1}. Name: {task['name']}, Priority: {task['priority_text']}")
    print("-------------------\n")
    return sorted_tasks # Return sorted tasks for other functions

def complete_task():
    if not tasks:
        print("No tasks to complete.\n")
        return

    sorted_tasks = list_tasks() # Display tasks and get the sorted list

    while True:
        try:
            task_number = int(input("Enter the number of the task to complete: "))
            if 1 <= task_number <= len(sorted_tasks):
                break
            else:
                print(f"Invalid task number. Please enter a number between 1 and {len(sorted_tasks)}.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    # The index in `tasks` might be different from `sorted_tasks`
    # Find the original task object to remove it from the 'tasks' list
    completed_task_obj = sorted_tasks[task_number - 1]

    # Remove the task from the main tasks list
    tasks.remove(completed_task_obj)

    # Save to file
    with open(COMPLETED_TASKS_FILE, 'a') as f:
        f.write(f"{completed_task_obj['name']}\n")

    print(f"Task '{completed_task_obj['name']}' completed and saved to {COMPLETED_TASKS_FILE}.\n")

def what_is_next():
    priority_1_tasks = [task for task in tasks if task['priority'] == 1]
    if priority_1_tasks:
        # Tasks are added to the 'tasks' list in chronological order,
        # so the first Priority 1 task found is the oldest.
        next_task = priority_1_tasks[0]
        print(f"Your next high-priority task is: '{next_task['name']}'.\n")
    else:
        print("Great job! You don't have any urgent Priority 1 tasks right now. Keep up the good work!\n")

def main():
    print("Welcome to your Personal Task Manager!")
    while True:
        print("Please choose an option:")
        print("1. Add a new task")
        print("2. List all tasks")
        print("3. Complete a task")
        print("4. What is next?")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ").strip()

        if choice == '1':
            add_task()
        elif choice == '2':
            list_tasks()
        elif choice == '3':
            complete_task()
        elif choice == '4':
            what_is_next()
        elif choice == '5':
            print("Exiting Task Manager. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter 1, 2, 3, 4, or 5.\n")

if __name__ == "__main__":
    main()
```

### How to Use the Task Manager:

1.  **Run the code cell** above.
2.  You will be presented with a menu:
    *   `1. Add a new task`
    *   `2. List all tasks`
    *   `3. Exit`
3.  **To add a task**: Enter `1`, then provide the task name and a priority level (1 for High, 2 for Medium, 3 for Low).
4.  **To view tasks**: Enter `2`. The tasks will be displayed, sorted by priority (High to Low).
5.  **To exit**: Enter `3`.
