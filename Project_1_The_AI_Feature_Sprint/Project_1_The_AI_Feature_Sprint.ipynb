```python
tasks = []

def add_task():
    task_name = input("Enter task name: ").strip()
    if not task_name:
        print("Task name cannot be empty. Please try again.")
        return

    while True:
        try:
            priority = int(input("Enter priority (1: High, 2: Medium, 3: Low): "))
            if 1 <= priority <= 3:
                break
            else:
                print("Priority must be 1, 2, or 3. Please try again.")
        except ValueError:
            print("Invalid input. Please enter a number for priority.")

    priority_map = {1: 'High', 2: 'Medium', 3: 'Low'}
    tasks.append({'name': task_name, 'priority': priority, 'priority_text': priority_map[priority]})
    print(f"Task '{task_name}' with priority '{priority_map[priority]}' added.\n")

def list_tasks():
    if not tasks:
        print("No tasks to display.\n")
        return

    # Sort tasks by priority (lower number = higher priority)
    sorted_tasks = sorted(tasks, key=lambda x: x['priority'])

    print("--- Your Tasks ---")
    for i, task in enumerate(sorted_tasks):
        print(f"{i+1}. Name: {task['name']}, Priority: {task['priority_text']}")
    print("-------------------\n")

def main():
    print("Welcome to your Personal Task Manager!")
    while True:
        print("Please choose an option:")
        print("1. Add a new task")
        print("2. List all tasks")
        print("3. Exit")

        choice = input("Enter your choice (1-3): ").strip()

        if choice == '1':
            add_task()
        elif choice == '2':
            list_tasks()
        elif choice == '3':
            print("Exiting Task Manager. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter 1, 2, or 3.\n")

if __name__ == "__main__":
    main()
```

### How to Use the Task Manager:

1.  **Run the code cell** above.
2.  You will be presented with a menu:
    *   `1. Add a new task`
    *   `2. List all tasks`
    *   `3. Exit`
3.  **To add a task**: Enter `1`, then provide the task name and a priority level (1 for High, 2 for Medium, 3 for Low).
4.  **To view tasks**: Enter `2`. The tasks will be displayed, sorted by priority (High to Low).
5.  **To exit**: Enter `3`.
