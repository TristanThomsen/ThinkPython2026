```python
###### ROSTER TEAM PROGRAM ######
# Need to be able to view the roster
# Need to be able to add a player to the roster
# Need to be able to remove a player from the roster
# Need to be able to "Quit!" when inputed (user input's "Quit" to leave menu)

###### IMPORT STATEMENTS ######
# None
###### FUNCTION DEFINITONS ######

# 1. View the Roster of players.
def view_roster(roster):
  if not roster:
    print("Roster is currently empty.")
  else:
    print("\nCurrent Roster:")
    for player in roster:
      print(f"- {player}")

# 2. Add players to the roster.
def add_player(roster, player_name):
  if player_name in roster:
    print(f"'{player_name}' is already on the roster.")
  else:
    roster.append(player_name)
    print(f"'{player_name}' has been added to the roster.")
  view_roster(roster)

# 3. Remove the players from the roster.
def remove_player(roster, player_name):
  if player_name in roster:
    roster.remove(player_name)
    print(f"'{player_name}' has been removed from the roster.")
  else:
    print(f"'{player_name}' is not on the roster.")
  view_roster(roster)

# 4. Edit a Players Name on the roster.
def edit_roster(roster, old_player_name, new_player_name):
  if old_player_name in roster:
    index = roster.index(old_player_name)
    roster[index] = new_player_name
    print(f"'{old_player_name}' has been changed to '{new_player_name}'.")
  else:
    print(f"'{old_player_name}' is not on the roster.")
  view_roster(roster)

# 5. Check if a player is on the roster.
def is_on_roster(roster, player_name):
  return player_name in roster

###### PROGRAM LOGIC ######

# Create the empty roster
roster = []

# While True Loop
while True:
  # Roster Menu appears with a list to choose from.
  try:
    user_option = int(input("""
          Welcome to the Roster!
  Please select one of the following options:

        1. See the roster
        2. Add a player to the roster
        3. Remove a player from the roster
        4. Edit a player's name
        5. Check if a player is on the roster
        6. Quit

        Enter your choice: """))

    if user_option == 1:
      view_roster(roster)

    elif user_option == 2:
      player_name = input("What is the name of the player you would like to add to the roster? ")
      add_player(roster, player_name)

    elif user_option == 3:
      player_name = input("What is the name of the player you would like to remove from the roster? ")
      remove_player(roster, player_name)

    elif user_option == 4:
      if not roster:
        print("Roster is empty. Cannot edit players.")
        continue
      old_player_name = input("What is the name of the player you would like to edit? ")
      new_player_name = input("What is the new name of the player you would like to change to? ")
      edit_roster(roster, old_player_name, new_player_name)

    elif user_option == 5:
      player_name = input("What is the name of the player you would like to check? ")
      if is_on_roster(roster, player_name):
        print(f"'{player_name}' is on the roster.")
      else:
        print(f"'{player_name}' is NOT on the roster.")

    elif user_option == 6:
      print("Thanks for viewing the roster! Goodbye!")
      break

    else:
      print("Invalid option. Please choose a number between 1 and 6.")

  except ValueError:
    print("Invalid input. Please enter a number.")
  except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
